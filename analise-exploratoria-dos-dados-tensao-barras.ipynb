# ANÁLISE EXPLORATÓRIA DE DADOS - BARRAS

```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from datetime import datetime
#registro de converters para uso do matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Carregamento da base de dados
base = pd.read_csv('dataset-tensao-barras-mes-jan-2021.csv', skiprows=0, sep = ';')
base.head(5)
```

```python
# DESCRIÇÃO COMPLETA DOS DADOS
base.describe(include="all")
```

```python
# DIMENSÕES DO DATASET
base.shape
```

```python
# Visualização do tipo de dados dos atributos
print(base.dtypes)
```

```python
#Transformar campos para o tipo de dados correto
base['data'] = pd.to_datetime(base['data'])
base['data_v2'] = pd.to_datetime(base['data_v2'])
base['data_hora'] = pd.to_datetime(base['data_hora'])
base['data_hora_v2'] = pd.to_datetime(base['data_hora_v2'])
```

```python
# Visualização do tipo de dados dos atributos
print(base.dtypes)
```

```python
# Visualização do índice do dataframe 
base.index
```

```python
#Obter o array de barras
df_aux = base
df_barras = base['barra'].unique()
df_barras
```

```python
#Criar range de dados
n_datas=5
rng_datas = pd.date_range('1/1/2021', periods = n_datas, freq = 'D')
rng_datas_minutos = pd.date_range('1/1/2021', periods = n_datas*1440, freq = 'T')

rng_datas
rng_datas_minutos
```

```python
# Exibir o gráfico de 01 barras em 01 dia
str_barra = 'MRBP4-01'
dt_data = '2021-01-27'

df = base
df = df_aux.loc[(df_aux.barra == str_barra) & ( df_aux.data == dt_data )  ]
df = df[['horario','lim_inf','lim_recom','lim_sup','valor_tensao'] ]
df.set_index("horario",  inplace=True)

fig = df.plot(); 
fig.set_ylabel('Tensão (kV)')
fig.set_xlabel('Horários ' + str(dt_data) )
fig.set_title('Tensão na Barra ' + str_barra + ' em ' + dt_data)

plt.legend(loc = 'best')
```

```python
# Exibir o gráfico de 01 barras em vários dias seguidos
str_barra = 'LIBP4-01'
dt_data_ini = '2021-01-01'
dt_data_fim = '2021-01-07'

df = base
df = df_aux.loc[(df_aux.barra == str_barra) & ( df_aux.data >= dt_data_ini ) &  (df_aux.data <= dt_data_fim)]
df = df[['horario','lim_inf','lim_recom','lim_sup','valor_tensao'] ]

df.set_index("horario",  inplace=True)

fig = df.plot(); 
fig.set_ylabel('Tensão (kV)')
fig.set_xlabel('Horários ' + str(dt_data) )
fig.set_title('Tensão na Barra ' + str_barra + ' em ' + dt_data)

plt.legend(loc = 'best')
```

```python
i=0
j=0
n_datas=4
str_barra = 'PFBP4-01'
rng_datas = pd.date_range('1/1/2021', periods = n_datas, freq = 'D')

fig, ax = plt.subplots(nrows = 1, ncols = n_datas, figsize=(15,5))

for dt_data in rng_datas:
        
    print("barra: " + str_barra + " || data: " + str(dt_data))
        
    df = df_aux.loc[(df_aux.barra == str_barra) & ( df_aux.data == dt_data ) ]
    df = df[['horario', 'lim_inf','lim_recom','lim_sup','valor_tensao'] ]
    df.set_index("horario",  inplace=True)
    
    index = dt_data.day-1
    ax[index].set_ylabel('Tensão (kV)')
    #ax[index].set_xlabel(str(dt_data) )
    ax[index].set_title(str_barra + ' / ' + str(dt_data).split()[0])            
    ax[index].plot(df)

#fig.legend()
#fig.show()       
```

```python
# Exibir o gráfico de várias barras em vários dias

i = 1
j = 1
k = 0

vetor_cores = ['','','','','']

n_barras = 5
n_dt_datas = 5

# Gráficos em Paralelo
fig, axes = plt.subplots(nrows = n_barras, ncols = n_dt_datas)

for str_barra in df_barras:
    
    for dt_data in rng_datas:
        
        print("barra: " + str_barra + " || data: " + dt_data)
        
        df = df_aux.loc[(df_aux.barra == str_barra) & ( df_aux.data = dt_data ) ]
        df = df[['lim_inf','lim_recom','lim_sup','valor_tensao'] ]
        
        axes[i,j].set_xlabel('Data ' + dt_data)
        axes[i,j].set_ylabel('Tensão (kV)' )
        axes[i,j].set_title('Tensão na Barra ' + str_barra + ' em ' + dt_data);
        
        axes[i,j].plot(df)

#plt.figure(); 
#df.plot(); 
fig.legend(loc = 'best')
fig.show()        
        
```
